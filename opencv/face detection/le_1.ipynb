```python
import cv2 as cv
from PIL import Image
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
import os
from tqdm import tqdm


BASE_DATA_DIR = "../data"
CELEB = "/celebrities"
VIDEOS = "/videos"

def display_as_rgb(img):
  to_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
  display(Image.fromarray(to_rgb))

os.listdir(BASE_DATA_DIR + CELEB)
```

these face detection models usually use edges to try to determine whether an object is a face or not

```python
img = cv.imread(BASE_DATA_DIR + CELEB + "/celeb-1.png")
display_as_rgb(img)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
display_as_rgb(gray)
```

haar cascade is a model that can be used to detect faces, keep in mind that it is only ideal in simple projects.

```python
# load face detection model
model_path = os.getcwd() + "/model.xml"
haar_cascade = cv.CascadeClassifier(model_path)
```

```python
# detect faces in the gray image
faces_rect = haar_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3)
print(f"number of faces found: {len(faces_rect)}")
```

```python
# highlight the detected faces
for (x, y, w, h) in faces_rect:
  cv.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), thickness=2)

display_as_rgb(img)
```
