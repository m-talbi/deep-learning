Bitwise operations are used a lot in image processing especially when working with masks

```python
blank = np.zeros((400, 400), dtype=np.uint8)
```

```python
def display_as_rgb(img):
  to_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
  display(Image.fromarray(to_rgb))
```

```python
rectangle = cv.rectangle(blank.copy(), (30, 30), (370, 370), 255, -1)

to_rgb = cv.cvtColor(rectangle, cv.COLOR_BGR2RGB)
display(Image.fromarray(to_rgb))
```

```python
circle = cv.circle(blank.copy(), (200, 200), 200, 255, -1)

to_rgb = cv.cvtColor(circle, cv.COLOR_BGR2RGB)
display(Image.fromarray(to_rgb))
```

```python
# bitwise AND
# returns the intersection between 2 images
bitwise_and = cv.bitwise_and(rectangle, circle)
display_as_rgb(bitwise_and)
```

```python
# bitwise OR
# returns the intersecting and the non-intersecting regions
# basically merges the two images
bitwise_or = cv.bitwise_or(rectangle, circle)
display_as_rgb(bitwise_or)
```

```python
# bitwise XOR
# returns the non-intersecting regions
bitwise_xor = cv.bitwise_xor(rectangle, circle)
display_as_rgb(bitwise_xor)
```

```python
# bitwise NOT
# inverts the binary color
bitwise_not = cv.bitwise_not(rectangle)

# black becomes white and white becomes black
display_as_rgb(bitwise_not)
display_as_rgb(cv.bitwise_not(circle))
```
