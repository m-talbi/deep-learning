```python
import cv2 as cv
from PIL import Image
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
import os


BASE_DATA_DIR = "../data"
CATS = "/cat images"
VIDEOS = "/videos"

def display_as_rgb(img):
  to_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
  display(Image.fromarray(to_rgb))

os.listdir(BASE_DATA_DIR + CATS)
```

From mathematical POV, gradients and edges are completely different things, but from programming perspective you can think of them as the same thing.

```python
img = cv.imread(BASE_DATA_DIR + CATS + "/cat-5.png")
display_as_rgb(img)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
display_as_rgb(gray)
```

```python
# Laplacian method
lap = cv.Laplacian(gray, cv.CV_64F)
lap = np.uint8(np.absolute(lap))
display_as_rgb(lap)
```

```python
img2 = cv.imread(BASE_DATA_DIR + CATS + "/cat-1.png")
gray2 = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)
lap2 = cv.Laplacian(gray2, cv.CV_64F)
lap2 = np.uint8(np.absolute(lap2))
display_as_rgb(lap2)
```

```python
# Sobel method
sobel_x = cv.Sobel(gray, cv.CV_64F, 1, 0)
sobel_y = cv.Sobel(gray, cv.CV_64F, 0, 1)
display_as_rgb(sobel_x.astype(np.uint8))
display_as_rgb(sobel_y.astype(np.uint8))
```

```python
combined = cv.bitwise_or(src1=sobel_x, src2=sobel_y)
display_as_rgb(combined.astype(np.uint8))
```

```python
# Canny method
canny = cv.Canny(gray, 150, 175)
display_as_rgb(canny)
```
