```python
import cv2 as cv
from PIL import Image
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
import os


BASE_DATA_DIR = "../data"
CATS = "/cat images"
VIDEOS = "/videos"

def display_as_rgb(img):
  to_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
  display(Image.fromarray(to_rgb))

os.listdir(BASE_DATA_DIR + CATS)
```

#### Thresholding

Thresholding is a binary realization of an image. In general we want to take an image and convert it to binary image that is where pixels are either black ir white.

```python
img = cv.imread(BASE_DATA_DIR + CATS + "/cat-2.png")
display_as_rgb(img)
```

```python
# Simple thresholding
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
display_as_rgb(gray)
```

```python
# thresh: if the intensity of the pixel is higher 
# than the thresh value, it will be set to maxval
threshold, thresh = cv.threshold(gray, thresh=150, maxval=255, type=cv.THRESH_BINARY)
display_as_rgb(thresh)
```

```python
# the inverse of the previous image
threshold, thresh_inv = cv.threshold(gray, thresh=150, maxval=255, type=cv.THRESH_BINARY_INV)
display_as_rgb(thresh_inv)
```

```python
# Adaptive threshold
# blockSize: is the neighborhood size of the kernel size, used to compute the mean
# C: an integer subtracted from the mean, allows to fine tune the threshold
adaptive_thresh = cv.adaptiveThreshold(src=gray, maxValue=255, adaptiveMethod=cv.ADAPTIVE_THRESH_MEAN_C, thresholdType=cv.THRESH_BINARY, blockSize=11, C=3)
display_as_rgb(adaptive_thresh)
```

```python
# Inverse
adaptive_thresh2 = cv.adaptiveThreshold(src=gray, maxValue=255, adaptiveMethod=cv.ADAPTIVE_THRESH_MEAN_C, thresholdType=cv.THRESH_BINARY_INV, blockSize=11, C=3)
display_as_rgb(adaptive_thresh2)
```

```python
# Tuning C value
adaptive_thresh2 = cv.adaptiveThreshold(src=gray, maxValue=255, adaptiveMethod=cv.ADAPTIVE_THRESH_MEAN_C, thresholdType=cv.THRESH_BINARY, blockSize=11, C=5)
display_as_rgb(adaptive_thresh2)
```

```python
# Gaussian method
adaptive_thresh2 = cv.adaptiveThreshold(src=gray, maxValue=255, adaptiveMethod=cv.ADAPTIVE_THRESH_GAUSSIAN_C, thresholdType=cv.THRESH_BINARY, blockSize=11, C=7)
display_as_rgb(adaptive_thresh2)
```
