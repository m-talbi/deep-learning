```python
import cv2 as cv
from PIL import Image
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
import os


BASE_DATA_DIR = "../data"
CATS = "/cat images"
VIDEOS = "/videos"

os.listdir(BASE_DATA_DIR + CATS)
```

```python
img = cv.imread(BASE_DATA_DIR + CATS + "/cat-4.png")
to_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
display(Image.fromarray(to_rgb))
```

```python
# RGB, BGR, grayscale, HSV are all color spaces
# grayscale shows the distribution of pixel intensity

# BGR to HSV
hsv = cv.cvtColor(img, code=cv.COLOR_BGR2HSV)
cv.imshow("hsv", hsv)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# BGR to LAB

lab = cv.cvtColor(img, code=cv.COLOR_BGR2LAB)
cv.imshow("lab", lab)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Color channels (merge and split)
b, g, r = cv.split(img)
```

```python
# Blue
display(Image.fromarray(b))
```

```python
# Green
display(Image.fromarray(g))
```

```python
# Red
display(Image.fromarray(r))
```

```python
# The more whiter the image/spot the higher the intensity of that color
# in the cat example, red has a high intensity
```

```python
# merge colors

merged = cv.merge([b, g, r])
rgb = cv.cvtColor(merged, cv.COLOR_BGR2RGB)
display(Image.fromarray(rgb))
```

```python
blank = np.zeros(img.shape[:2], dtype=np.uint8)

b, g, r = cv.split(img)

blue = cv.merge([b, blank, blank])
green = cv.merge([blank, g, blank])
red = cv.merge([blank, blank, r])
```

```python
# Blue (image.fromarray uses RGB) blue -> red
display(Image.fromarray(red))
```

```python
# Green
display(Image.fromarray(green))
```

```python
# Red (image.fromarray uses RGB) Red -> blue
display(Image.fromarray(blue))
```
