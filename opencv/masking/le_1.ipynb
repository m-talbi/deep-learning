```python
import cv2 as cv
from PIL import Image
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
import os


BASE_DATA_DIR = "../data"
CATS = "/cat images"
VIDEOS = "/videos"

def display_as_rgb(img):
  to_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
  display(Image.fromarray(to_rgb))

os.listdir(BASE_DATA_DIR + CATS)
```

Masking allows us to focus on certain parts of the image that we want to work on.

```python
img = cv.imread(BASE_DATA_DIR + CATS + "/cat-4.png")
display_as_rgb(img)
```

```python
blank = np.zeros(img.shape[:2], dtype=np.uint8)
display_as_rgb(blank)
```

```python
# you can control the position of the circle by adjusting the center parameter
mask = cv.circle(blank.copy(), (img.shape[1] // 3 + 10, img.shape[0] // 3 + 25), 75, 255, -1)
display_as_rgb(mask)
```

```python
masked = cv.bitwise_and(img, img, mask=mask)
display_as_rgb(masked)
```
