##### Gradient Descent Algorithm
Gradient descent algorithm does not work for all functions. There are two specific requirements. A function has to be:

- differentiable
- convex

#### Convex

for a function to be convex. For a univariate function, this means that the line segment connecting two functions points `lays on or above its curve (it does not cross it)`. If it does it means that it has a local minimum which is not a global one.



Mathematically, for two points x, x laying on the functions curve this condition is expressed as:

where  denotes a points location on a section line and its value has to be between 0 (left point) and 1 (right point), e.g. =0.5 means a location in the middle.

$$f(\lambda x_1 + (1 - \lambda)x_2) \leq \lambda f(x_1) + (1 - \lambda) f(x_2)$$

is the definition of a convex function.

Explanation:

A function $f: \mathbb{R}^n \to \mathbb{R}$ is called convex if, for all $x_1, x_2 \in \mathbb{R}^n$ and all $\lambda \in [0, 1]$, the inequality

$$f(\lambda x_1 + (1 - \lambda) x_2) \leq \lambda f(x_1) + (1 - \lambda) f(x_2)$$

holds.

Another way to check mathematically if a univariate function is convex is to calculate the second derivative and check if its value is always bigger than 0.

$$\frac{d^2 f}{d x^2} > 0$$
